## 1. The Stakeholder

**Larry Sanders - 42 years old, married, 2 kids**

He wants to buy a house for him and his family in Kings County, Washingtion USA. 

He listed the following **criteria** for his new house:
1) The house should have a waterfront view.
2) Because of his germ phobia, the house should be isolated, but still central (ideally without many kids around)
3) Since he is going to live there with his kids, the house needs enough space, e.g. bathrooms and bedrooms.

However, Larry has a limited **budget**. He cannot afford houses with pricing higher than 1.5 M$.

## 2. Aims and Scope

I need to recommend a set of houses to Larry which best fit his criteria. At the same time, I need to take into consideration limiting factors, particularly his limited budget.

I will perform exploratory data analysis on the available houses in King County to select and recommend those houses, which are closest to Larry´s expectations. To this end, I first derive research questions and hypotheses to guide the analysis strategy.

Data analysis will be performed depending on the following stakeholder criteria: 
* availability of waterfront views, centralized location, number of bed- and bathrooms, housing price

## 3. Research Questions and Hypotheses

**1) What is the availability and price of houses with a waterfront view?**
    
* Houses with waterfront are less common than those without.
* Houses with waterfront are more expensive than those without.

**2) What is the price of houses in central, but isolated neighborhoods?**

* Houses in central, but isolated neighborhoods are more expensive than at other sites.

**3) How does the number of bed- and bathrooms affect housing price?**

## 5. Exploratory data analysis


### Q1: What is the availability and price of houses with a waterfront view?

##### H1.1: Houses with waterfront are less common than those without.

First, I will determine the availability of houses with a waterfront view. To this end, I will count the number of houses with and without waterfront and plot them in a countplot. 

Additionally, I will also count and print the raw number of houses without and without waterfront as well as of those without this information.

```python
plt.figure(figsize=(5,3))

sns.countplot(data=df, x='waterfront', 
                alpha=0.5, edgecolor='black', palette={0:'blue', 1: 'red'})
plt.ylabel('Number of houses [n]', fontsize=11)
plt.xlabel('Waterfront View', fontsize=11)
plt.xticks(ticks=np.arange(0, 2), labels=['no', 'yes'])
plt.title('Availability of Houses depending on Waterfront View\n')

plt.show()

print(f'There are {df.waterfront.value_counts()[0]} houses without and only {df.waterfront.value_counts()[1]} houses with waterfront view.')
print('Additionally, information on waterfront view is missing for', 
        df.waterfront.isna().sum(), 'houses')
```

The countplot depicts the number of houses with ('yes', blue) and without waterfront view ('no', red). As can be seen, there are only about 150 houses with waterfront view. Thus, these data confirm my first hypothesis and show that there are only few houses meeting the stakeholder´s first criterion.

Of note, information on waterfront view is missing for a total of 2.376 houses.

#### H1.2: Houses with waterfront are more expensive than those without.

Since the stakeholder has a limited budget, I next examined price differences between houses with and without a waterfront. To this end, I plotted the median housing price and the distribution of prices and compared them between the two conditions. I chose the median since the initial data exploration showed showed that price values are not normally distributed and strongly skewed to the right. The distribution of prices was plotted in the form of a histogram.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 4), gridspec_kw={'width_ratios': [1, 3]})
fig.suptitle('Housing Prices by Waterfront View', fontsize=14)

# median housing price
fig1 = sns.barplot(data=df, x='waterfront', y='price', alpha=0.5, edgecolor='black',
        errwidth=0, palette={0:'blue', 1: 'red'}, ax=ax1)
ax1.set_ylabel('Housing Prices [M$]', fontsize=11)
ax1.set_yticks(ticks=np.arange(0, 2500000, 500000), labels=np.arange(0, 2.5, 0.5).round(1))
ax1.set_xlabel('Waterfront View', fontsize=11)
ax1.set_xticks(ticks=np.arange(0, 2), labels=['no', 'yes'])
ax1.set_title('Median')

# relative distribution of housing prices
fig2 = sns.histplot(data=df[df.waterfront==0], x='price', bins=50, stat='probability',
        color='blue', alpha=0.5, label='no', ax=ax2)
fig3 = sns.histplot(data=df[df.waterfront==1], x='price', bins=50, stat='probability',
        color='red', alpha=0.5, label='yes', ax=ax2)
ax2.set_xlabel('Housing Prices [M$]', fontsize=11)
ax2.set_xticks(ticks=np.arange(0, 9000000, 1000000), labels=np.arange(0, 9))
ax2.set_ylabel('Proportion of Houses', fontsize=12)
ax2.legend(title='Waterfront')
ax2.set_title('Relative Distribution', fontsize=12)

plt.tight_layout()
plt.show()
```

The figure above depicts price differences between houses with ('yes', red) and without waterfront ('no', blue). The left-hand panel depicts the median housing price. The right-hand panel depicts the distribution of prices as a histogram, with data being binned into 50 bins. Bins are displayed as proportions relative to the total number of houses per condition, since the total number of houses with and without waterfront view strongly differed.

As can be seen from the figure, houses with waterfront view are considerably more expensive than those without, which is reflected both in a higher median and a shifted price distribution towards extreme prices in the million range, thus confirming my second hypothesis. However, some houses with waterfront view are still within the stakeholder´s budget limit of 1.5 M$.

Since information on waterfront view is missing for more than 2000 houses, I additionally analyzed the relationship between geographical position and housing price.

I therefore next plotted the relationship between housing price and geographical position using a geospatial scatterplot, with longitude and latitude as serving x- and y-coordinates for house locations. Houses were then color-coded by their price.

```python
# using matplotlib
fig = plt.scatter(data=df, x='long', y='lat', c='price', s=1, 
        cmap='seismic', norm=colors.LogNorm(df.price.min(), df.price.max()))
cbar = plt.colorbar(fig, label='Price [M$]')
plt.xlabel('Longitude', fontsize=11)
plt.ylabel('Latitude', fontsize=11)
plt.title('Housing prices depending on geographical position')
plt.show()
```

```python
# using plotly.express
fig = px.scatter_mapbox(df, lat='lat', lon='long', hover_name='id',
                        hover_data=['zipcode','price'], color=np.log10(df.price), # use log-transformed prices
                        color_continuous_scale=px.colors.sequential.Bluered,
                        zoom=8, height=800, opacity=0.5)

fig.update_traces(marker=dict(size=4, opacity=0.7),
                selector=dict(mode='markers'))

fig.layout.coloraxis.colorbar.title = 'log10(Price)'

fig.update_layout(mapbox_style='open-street-map',
                    height=500, margin={'r':0,'t':30,'l':0, 'b':0}, # sets the margin from the left, top, right and bottom order
                    title_text='Houses in King County color-coded by Price', title_x=0.5) # sets a title for the plot and centers it

fig.show()
```

The two figures above shows a geospatial scatterplots of houses color-coded by their price. As can be seen, houses located at the coastline or near lakes and rivers are typically more expensive than those, which are not, again confirming my hypothesis.

### Q2: What is the price of houses in centralized, isolated neighborhoods?

#### H2.1: Houses in centralized, isolated neighborhoods are more expensive than at other sites.


The stakeholder prefers a house with a central location but in an isolated neighborhoods. There ware no direct metrics in the dataset. 

Thus, I first generated a readout for centrally located houses. Cities in King County are typically located at the coastline. I therefore checked again the geographical distribution of houses on a geospatial map in order to extract longitude and latitude criteria for defining houses with and without central locations. 

```python
fig = sns.scatterplot(data=df, x='long', y='lat', hue='zipcode', s=2)
plt.xlabel('Longitude', fontsize=11)
plt.ylabel('Latitude', fontsize=11)
plt.title('Geogaphrical Distribution of Houses color-coded by Zipcode')
plt.show()
```

The figure above illustrates the geospatial distribution of houses according to longitude and latitude, with houses color-coded by zipcode. With reference to the plotly-derived geospatial map above, it becomes apparent that houses with longitude values between -122.4 and -122.1 and latitude values between 47.4 and 47.8 are associated with city areas, in close proximity to the ocean or Lake Washington. Houses with those values can be considered centralized while the remainder can be considered not central.

Based on these values, i designed a new feature coding houses are located central or not.

```python
# compute new column based on longitude and latitude criteria defined above
df['centralized'] = np.where(((df.long <= -122.1) & (df.long >= -122.4) 
                        & (df.lat >= 47.4) & (df.lat <= 47.8)),
                        1, 0)
```

```python
# check if minimum values per condition are correct
df[['lat', 'long', 'centralized']].groupby('centralized').min()
```

```python
# check if maximum values per condition are correct
df[['lat', 'long', 'centralized']].groupby('centralized').max()
```

Due to his germ phobia, the stakeholder also wants to live in an isolated neighborhood, ideally with only few kids. Since dataset does not contain any direct measures, I again generated a readout by proxy.

The dataset contains the column 'sqft_lot15', which represents the square footage of the land lots of the 15 nearest neighbors, which can serve as a proxy for isolated neighborhoods. The bigger the square footage, the more dispersed are the 15 nearest neighbors, thus resulting in a lower likelihood of meeting them.

I therefore generated a new feature 'isolated' based on the 'sqft_lot15' column. I set the threshold for a house being considered isolated to the 75% quantile of the 'sqft_lot15' data, which roughly corresponds to a square footage of 10.000 square feet of the 15 nearest neighbors.

```python
# compute new column based on square footage of land lots of 15 nearest neighbors as defined above
df['isolated'] = np.where((df.sqft_lot15 > df.sqft_lot15.quantile(0.75)), 1, 0)
```

```python
# check if minimum values per condition are correct
df[['sqft_lot15', 'isolated']].groupby('isolated').min()
```

```python
# check if maximum values per condition are correct
df[['sqft_lot15', 'isolated']].groupby('isolated').max()
```

```python
# count houses which are centralized and isolated and which are not
df.groupby(['centralized', 'isolated']).size().to_frame('frequency')
```

There are still 2637 houses fulfilling both criteria. I therefore filtered the dataset for housing meeting up with the criterion of being isolated and then compared prices between houses with and without centralized locations.

```python
# select only houses in isolated neighborhood
isolated_houses = df[df.isolated == 1]
```

Next, I plotted to prices between houses of in isolated neighborhoods with and without central location. Because of the skewness of the price data, we plotted both the median value and the distribution of prices for those houses.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 4), gridspec_kw={'width_ratios': [1, 3]})
fig.suptitle('Prices of Houses in Isolated Neighborhood depending on Location', fontsize=14)

# median housing price
fig1 = sns.barplot(data=isolated_houses, x='centralized', y='price', alpha=0.5, edgecolor='black',
        errwidth=0, palette={0:'grey', 1: 'green'}, ax=ax1)
ax1.set_ylabel('Housing Prices [M$]', fontsize=11)
ax1.set_yticks(ticks=np.arange(0, 2500000, 500000), labels=np.arange(0, 2.5, 0.5).round(1))
ax1.set_xlabel('centralized', fontsize=11)
ax1.set_xticks(ticks=np.arange(0, 2), labels=['no', 'yes'])
ax1.set_title('Median')

# relative distribution of housing prices
fig2 = sns.histplot(data=isolated_houses[isolated_houses.centralized==0], 
        x='price', bins=50, stat='probability',
        color='grey', alpha=0.5, label='no', ax=ax2)
fig3 = sns.histplot(data=isolated_houses[isolated_houses.centralized==1], 
        x='price', bins=50, stat='probability',
        color='green', alpha=0.5, label='yes', ax=ax2)
ax2.set_xlabel('Housing Prices [M$]', fontsize=11)
ax2.set_xticks(ticks=np.arange(0, 9000000, 1000000), labels=np.arange(0, 9))
ax2.set_ylabel('Proportion of Houses', fontsize=12)
ax2.legend(title='centralized')
ax2.set_title('Relative Distribution', fontsize=12)

plt.tight_layout()
plt.show()
```

The figure above depicts pricing differences between houses in isolated neighborhoods with ('yes', green) and without centralized location ('no', grey). The left-hand panel depicts the median housing price. The right-hand panel depicts the distribution of prices as a histogram, with data being binned in 50 bins. Bins are displayed as proportions relative to the total number of houses per condition.

As can be seen, houses with central location are slightly more expensive than those without. Specifically, houses with central location have a slightly higher median price and slightly right-shifted price distribution. These results therefore confirm my initial hypothesis. However, there are still many houses in centralized, isolated neighborhoods within the stakeholder's budget limit. However, it should be considered that the centrality and isolation of houses could only be approximated. Additional information might be collected for more robust conclusions.

### Q3: How does the number of bed- and bathrooms affect housing price?

#### H3.1: Housing prices will increase with increasing number of bed- and bathrooms.

Since the stakeholder has kids, he houses to be recommended to the stakeholder need to have enough space to accommodate his family. Thus, I analyzed the relationship of the number of bedrooms and bathrooms with housing price to examine whether there are houses with enough space within the stakeholder´s budget limit. 

To this end, I plotted the relationship between these features and housing price in the form of scatterplots. Data was plotted separately for houses with and without waterfront views as well as for houses in central, isolated neighborhoods in order to assess whether there are houses available that meet up with the stakeholder´s other criteria besides room number.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
plt.suptitle('Change in Price Depending on Bedroom Number', fontsize=14)

ax1.scatter(x=df[df.waterfront==0].bedrooms-0.2, y=df[df.waterfront==0].price, alpha=0.5,
            color='blue', label='no')
ax1.scatter(x=df[df.waterfront==1].bedrooms+0.2, y=df[df.waterfront==1].price, alpha=0.5,
            color='red', label='yes')
ax1.set_ylabel('Housing Prices [M$]', fontsize=11)
ax1.set_yticks(ticks=np.arange(0, 8000000, 2000000), labels=np.arange(0, 8, 2))
ax1.set_xlabel('Bedrooms [n]', fontsize=11)
ax1.set_xticks(ticks=np.arange(0, 11, 1))
ax1.set_xlim(0, 10)
ax1.legend(title='Waterfront', loc='upper left')
ax1.set_title('Houses with or without Waterfront', fontsize=12)

ax2.scatter(x=df[df.centralized==0].bedrooms-0.2, y=df[df.centralized==0].price, alpha=0.5,
            color='grey', label='no')
ax2.scatter(x=df[df.centralized==1].bedrooms+0.2, y=df[df.centralized==1].price, alpha=0.5,
            color='green', label='yes')
ax2.set_ylabel('Housing Prices [M$]', fontsize=11)
ax2.set_yticks(ticks=np.arange(0, 8000000, 2000000), labels=np.arange(0, 8, 2))
ax2.set_xlabel('Bedrooms [n]', fontsize=11)
ax2.set_xticks(ticks=np.arange(0, 11, 1))
ax2.set_xlim(0, 10)
ax2.legend(title='centralized', loc='upper left')
ax2.set_title('Isolated Houses only by Location', fontsize=12)

plt.tight_layout()
plt.show()
```

The figure shows the relationship between the number of bedrooms and housing price. In the upper panel, houses are color-coded color-coded by having ('yes', red) or not having a waterfront ('no', blue). In the lower panel, houses are color-coded depending on whether they are located centrally ('yes', green) or not ('no', grey).

As can be seen, there is a near-linear increase in housing price with bedroom number up to about 5 bedrooms. For houses without waterfront view and those with central location, housing prices drop again with further increases in bedroom number, illustrating a curvilinear relationship, possibly due to a lower demand for houses with many bedrooms. Houses with waterfront views only have a maximum of 6 bedrooms, up to which only a near-linear increase is evident. 

For houses with waterfront view, there are also several cases with up to 4 bedrooms within the stakeholder´s budget limit of 1.5 M$. For houses with centralized locations, there are even examples with up to 7 bedrooms, which would be affordable by the stakeholder.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
plt.suptitle('Change in Price Depending on Bathroom Number', fontsize=14)

ax1.scatter(x=df[df.waterfront==0].bathrooms-0.2, y=df[df.waterfront==0].price, alpha=0.5,
            color='blue', label='no')
ax1.scatter(x=df[df.waterfront==1].bathrooms+0.2, y=df[df.waterfront==1].price, alpha=0.5,
            color='red', label='yes')
ax1.set_ylabel('Housing Prices [M$]', fontsize=11)
ax1.set_yticks(ticks=np.arange(0, 8000000, 2000000), labels=np.arange(0, 8, 2))
ax1.set_xlabel('Bathrooms [n]', fontsize=11)
ax1.set_xticks(ticks=np.arange(0, 7, 1))
ax1.set_xlim(0, 6)
ax1.legend(title='Waterfront', loc='upper left')
ax1.set_title('Houses with or without Waterfront', fontsize=12)

ax2.scatter(x=df[df.centralized==0].bathrooms-0.2, y=df[df.centralized==0].price, alpha=0.5,
            color='grey', label='no')
ax2.scatter(x=df[df.centralized==1].bathrooms+0.2, y=df[df.centralized==1].price, alpha=0.5,
            color='green', label='yes')
ax2.set_ylabel('Housing Prices [M$]', fontsize=11)
ax2.set_yticks(ticks=np.arange(0, 8000000, 2000000), labels=np.arange(0, 8, 2))
ax2.set_xlabel('Bathrooms [n]', fontsize=11)
ax2.set_xticks(ticks=np.arange(0, 7, 1))
ax2.set_xlim(0, 6)
ax2.legend(title='centralized', loc='upper left')
ax2.set_title('Isolated Houses only by Location', fontsize=12)

plt.tight_layout()
plt.show()
```

The figure shows the relationship between the number of bathrooms and housing price. In the upper panel, houses are color-coded by having ('yes', red) or not having a waterfront ('no', blue). In the lower panel, houses are color-coded depending on whether they are located centrally ('yes', green) or not ('no', grey).

As can be seen, there is a near-linear increase in housing price with bathroom number, confirming my hypothesis. However, a plateau seems to be reached in housing price with about 4 bathrooms. 

For houses with waterfront view, there are also several cases with 2 bathrooms within the stakeholder´s budget limit of 1.5 M$. The same holds true for houses with centralized locations.

### Recommendations

#### What houses can be recommended based on those criteria?

Finally, I filtered the dataset based on the stakeholder's criteria and limitations in order to extract a set of recommendable houses.

To get a first impression, I selected houses within the stakeholders budget limit (1.5 M$) and counted the occurrences of those houses fulfilling the criteria of being centralized but isolated and having a waterfront.

```python
# filter by budget limit, group by centralized, siolated and waterfront, and count occurrences
df[df.price < 1500000].groupby(['centralized', 'isolated', 'waterfront']).size().to_frame('frequency')
```

As can be seen from the table, there are indeed 27 houses within his price range and fulfilling all three other critera.

Since the house needs to accommodate this familiy, I further set a lower boundary for the number of bed- and bathrooms. Houses were additionally selected for at least three bedrooms (2 for his kids and 1 for himself and his wife) and at least one bathroom.

```python

df.loc[(df.price < 1500000) & (df.bedrooms >= 3) & (df.bathrooms >= 1)].groupby(['centralized', 'isolated', 'waterfront']).size().to_frame('frequency')
```

As can be seen in the table, there are still 17 houses meeting all of his criteria.

I generated a final, filtered dataset of houses that can be recommended to the stakeholder, including additional relevant information for further screening, e.g. the condition of the house, the price, the actual number of bed- and bathrooms, etc.

```python
recommended_houses = df.loc[(df.price < 1500000) & (df.waterfront==1) & (df.centralized==1) &
                    (df.isolated==1) & (df.bedrooms >= 3) & (df.bathrooms >= 1)]
recommended_houses = recommended_houses[['id', 'price', 'condition', 'bedrooms', 'bathrooms', 'sqft_living', 'yr_built', 'sqft_lot15']]
recommended_houses.sort_values(['price', 'condition'])
```
